```python
import numpy as np
import pandas as pd
from CAM import *
import models
import os
import os.path as osp
from utils.serialization import *
from utils.make_dir import *
import cv2
import torch
import PIL.Image as Image
from torchvision import transforms
```

```python
sp = r"D:\DATASETS\peso\train\CAM"
init_img = r"D:\DATASETS\peso\train\init_data_image"
check_p = r"D:\Code\TVCG\results\peso-result\epoch_25_model_best.pth.tar"
img_size = 224

#
model = models.create("resnet50", num_features=2048, num_classes=3)
checkpoint = torch.load(check_p)
model.load_state_dict(checkpoint['state_dict'], strict=False)
#
grad_cam = GradCam(model, target_layer_names=['base'], img_size=224)
#
for idx, pth in enumerate(os.listdir(init_img)):
    init_pth = os.path.join(init_img, pth)
    imgs = os.listdir(init_pth)
    for image in imgs:
        img = np.asarray(Image.open(os.path.join(init_pth, image)
                                    ).convert('RGB')).reshape((img_size, img_size, 3))
        img = np.float32(cv2.resize(img, (img_size, img_size))) / 255

        pre_img = preprocess_image(img)
        mask = grad_cam(pre_img)
        spp = os.path.join(sp, pth)
        make_dir(spp)
        show_cam_on_image(img, mask, os.path.join(spp, image))
```
